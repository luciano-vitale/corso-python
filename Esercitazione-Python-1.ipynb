**Esercizio:** scrivi un programma che chieda all'utente un numero intero positivo N e stampi tutti i numeri primi compresi tra 2 e N.

**Regole e vincoli:**


*   Un numero è primo se è divisibile solo per 1 e per se stesso.
*   Non puoi usare liste o altre strutture dati.

*   Devi usare un ciclo dentro l'altro (nested loops).
