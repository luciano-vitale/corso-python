## 1. Operatori relazionali (confronto)

### Cosa sono gli operatori relazionali?

Gli **operatori relazionali** (o **operatori di confronto**) permettono di confrontare due valori. Il risultato è sempre un valore **booleano**: `True` o `False`.

### 1.1 Tabella degli operatori di confronto

| Operatore | Significato | Esempio | Risultato |
|-----------|-------------|---------|-----------|
| `==` | Uguale a | `5 == 5` | `True` |
| `!=` | Diverso da | `5 != 3` | `True` |
| `>` | Maggiore di | `10 > 5` | `True` |
| `<` | Minore di | `3 < 7` | `True` |
| `>=` | Maggiore o uguale | `5 >= 5` | `True` |
| `<=` | Minore o uguale | `4 <= 6` | `True` |

### 1.2 Operatore == (uguale)

**Attenzione**: `==` (confronto) è diverso da `=` (assegnazione)!

````python
# == confronta due valori
print(5 == 5)           # True
print(5 == 3)           # False
print("ciao" == "ciao") # True
print("ciao" == "Ciao") # False (case-sensitive!)

# = assegna un valore
x = 5                   # Assegnazione
print(x == 5)           # True (confronto)
````

### 1.3 Operatore != (diverso)

````python
print(5 != 3)           # True
print(5 != 5)           # False
print("gatto" != "cane") # True
````

### 1.4 Operatore > (maggiore)

````python
print(10 > 5)           # True
print(5 > 10)           # False
print(5 > 5)            # False (non è maggiore, è uguale!)
````

### 1.5 Operatore < (minore)


````python
print(3 < 7)            # True
print(7 < 3)            # False
print(5 < 5)            # False
````

### 1.6 Operatore >= (maggiore o uguale)

````python
print(10 >= 5)          # True
print(5 >= 5)           # True (uguale va bene!)
print(3 >= 5)           # False
````

### 1.7 Operatore <= (minore o uguale)

````python
print(3 <= 7)           # True
print(5 <= 5)           # True
print(10 <= 5)          # False
````

## 2. La struttura `if`

### Cos'è una struttura condizionale?

Una **struttura condizionale** permette di eseguire del codice **solo se** una condizione è vera.

### 2.1 Sintassi di base

````python
if condizione:
    # codice da eseguire se condizione è True
    # IMPORTANTE: indentazione obbligatoria!
````

### Primo esempio


````python
eta = 18

if eta >= 18:
    print("Sei maggiorenne!")
    print("Puoi votare!")
````

**Nota importante**: L'**indentazione** (4 spazi) è **obbligatoria**! Definisce quale codice fa parte dell'`if`.

### Come funziona?

1. Python valuta la condizione dopo `if`
2. Se è `True`, esegue il blocco indentato
3. Se è `False`, salta il blocco

````python
# Esempio con True
if 5 > 3:
    print("Questo verrà stampato")  # Eseguito

# Esempio con False
if 5 < 3:
    print("Questo NON verrà stampato")  # Saltato
````

## 3. La struttura `if-else`

### Cos'è if-else?

`if-else` permette di eseguire un blocco di codice se la condizione è vera, e un **altro blocco** se è falsa.

### 3.1 Sintassi

```python
if condizione:
    # codice se condizione è True
else:
    # codice se condizione è False
```

### Esempio base

````python
eta = 16

if eta >= 18:
    print("Sei maggiorenne")
else:
    print("Sei minorenne")
````

### Come funziona?

- Se la condizione è `True`: esegue il blocco `if`
- Se la condizione è `False`: esegue il blocco `else`

**Uno dei due blocchi viene SEMPRE eseguito!**

````python
numero = 7

if numero % 2 == 0:
    print("Il numero è pari")
else:
    print("Il numero è dispari")
````

## 4. La struttura `if-elif-else`

### Cos'è elif?

`elif` (abbreviazione di "else if") permette di verificare **condizioni multiple**.

### 4.1 Sintassi

````python
if condizione1:
    # codice se condizione1 è True
elif condizione2:
    # codice se condizione2 è True
elif condizione3:
    # codice se condizione3 è True
else:
    # codice se nessuna condizione è True
````

### Esempio base


````python
voto = 85

if voto >= 90:
    print("Voto: Eccellente")
elif voto >= 75:
    print("Voto: Buono")
elif voto >= 60:
    print("Voto: Sufficiente")
else:
    print("Voto: Insufficiente")
````

### Come funziona?

1. Python controlla la prima condizione (`if`)
2. Se è `True`, esegue quel blocco e **finisce** (salta tutto il resto)
3. Se è `False`, passa alla prossima (`elif`)
4. Continua finché non trova una condizione `True`
5. Se nessuna è `True`, esegue `else` (se presente)

### Ordine importante!

```python
numero = 85

# CORRETTO - dal più specifico al più generale
if numero > 90:
    print("Eccellente")
elif numero > 75:
    print("Buono")      # Questo viene stampato
elif numero > 60:
    print("Sufficiente")
```

```python
numero = 85

# SBAGLIATO - ordine invertito
if numero > 60:
    print("Sufficiente")  # Questo viene stampato e si ferma!
elif numero > 75:
    print("Buono")        # Non viene mai raggiunto
elif numero > 90:
    print("Eccellente")   # Non viene mai raggiunto
```

## 5. Condizioni Annidate

### Cos'è l'annidamento?

Puoi mettere un `if` **dentro** un altro `if`. Questo si chiama **annidamento** (nesting).

### 5.1 Sintassi

```python
if condizione1:
    if condizione2:
        # codice se entrambe le condizioni sono True
```

### Esempio base

````python
eta = 20
ha_patente = True

if eta >= 18:
    print("Sei maggiorenne")
    if ha_patente:
        print("Puoi guidare!")
    else:
        print("Devi prendere la patente")
else:
    print("Sei minorenne")
````

### 5.2 Indentazione progressiva

Ogni livello di annidamento richiede **4 spazi in più**:

````python
numero = 15

if numero > 0:                          # Livello 0
    print("Positivo")                   # Livello 1 (4 spazi)
    if numero % 2 == 0:                 # Livello 1 (4 spazi)
        print("Pari")                   # Livello 2 (8 spazi)
    else:                               # Livello 1 (4 spazi)
        print("Dispari")                # Livello 2 (8 spazi)
else:                                   # Livello 0
    print("Negativo o zero")            # Livello 1 (4 spazi)
````

### Attenzione all'annidamento eccessivo!

Troppi livelli rendono il codice difficile da leggere:


```python

if condizione1:
    if condizione2:
        if condizione3:
            if condizione4:
                print("Troppo annidato!")
```

Spesso è meglio usare **operatori logici** (prossima sezione).

## 6. Operatori Logici

### Cosa sono gli operatori logici?

Gli **operatori logici** permettono di combinare più condizioni.

### I tre operatori logici

| Operatore | Significato | Esempio |
|-----------|-------------|---------|
| `and` | E (entrambe vere) | `x > 0 and x < 10` |
| `or` | O (almeno una vera) | `x < 0 or x > 10` |
| `not` | NON (negazione) | `not x == 5` |

### 6.1 Operatore `and`

Restituisce `True` **solo se entrambe** le condizioni sono vere:

````python
# Tabella verità di and
print(True and True)      # True
print(True and False)     # False
print(False and True)     # False
print(False and False)    # False

eta = 25
ha_patente = True

# Entrambe devono essere vere
if eta >= 18 and ha_patente:
    print("Puoi noleggiare un'auto")
````

### 6.2 Operatore `or`

Restituisce `True` se **almeno una** condizione è vera:

````python
# Tabella verità di or
print(True or True)       # True
print(True or False)      # True
print(False or True)      # True
print(False or False)     # False

giorno = "sabato"

# Basta che una sia vera
if giorno == "sabato" or giorno == "domenica":
    print("È weekend!")
````

### 6.3 Operatore `not`

Inverte il valore booleano:

````python
# Negazione
print(not True)           # False
print(not False)          # True

piove = False

if not piove:
    print("Possiamo uscire!")
````

### 6.6 Precedenza degli operatori logici

1. `not` (più alta)
2. `and`
3. `or` (più bassa)


````python
# Esempio
print(True or False and False)  # True
# Perché: and viene prima → False and False = False
#         poi: True or False = True

# Con parentesi per chiarezza
print((True or False) and False)  # False
````
