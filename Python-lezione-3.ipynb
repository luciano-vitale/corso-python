## 2. Operatori di base

Gli **operatori** sono simboli o parole chiave speciali che operano sui valori ed eseguono operazioni (matematiche).

### 2.1 Gli operatori aritmetici in Python

| Operatore | Nome | Esempio | Risultato |
|-----------|------|---------|-----------|
| `+` | Addizione | `5 + 3` | `8` |
| `-` | Sottrazione | `5 - 3` | `2` |
| `*` | Moltiplicazione | `5 * 3` | `15` |
| `/` | Divisione classica | `5 / 2` | `2.5` |
| `%` | Modulo (resto) | `5 % 2` | `1` |
| `**` | Esponente | `2 ** 3` | `8` |
| `//` | Divisione intera | `5 // 2` | `2` |

### 2.1.4 Divisione classica (/)

La divisione classica restituisce **sempre** un numero `float` (decimale):

```python
print(10 / 2)         # 5.0 (non 5!)
print(7 / 2)          # 3.5
print(5 / 2)          # 2.5
```

**Nota importante**: Anche quando il risultato è un numero intero, Python restituisce un float!

### 2.1.5 Divisione intera (//)

La divisione intera (o floor division) divide e arrotonda **verso il basso** al numero intero più vicino:

```python
print(7 // 2)         # 3 (non 3.5)
print(10 // 3)        # 3 (non 3.333...)
print(15 // 4)        # 3 (non 3.75)

# Funziona anche con i float
print(7.0 // 2.0)     # 3.0
print(10 // 3.0)      # 3.0
```

### 2.1.6 Operatore modulo (%)

L'operatore modulo restituisce il **resto** della divisione:

```python
print(10 % 3)         # 1 (perché 10 ÷ 3 = 3 resto 1)
print(14 % 4)         # 2 (perché 14 ÷ 4 = 3 resto 2)
print(15 % 5)         # 0 (perché 15 ÷ 5 = 3 resto 0)
print(7 % 2)          # 1 (per verificare se un numero è dispari)
print(8 % 2)          # 0 (per verificare se un numero è pari)
```

**Uso pratico**: L'operatore modulo è molto utile per verificare se un numero è pari o dispari!

### 2.1.7 Esponente (**)

L'operatore `**` eleva un numero a una potenza:

```python
print(2 ** 3)         # 8 (2 * 2 * 2)
print(3 ** 2)         # 9 (3 * 3)
print(5 ** 2)         # 25 (5 * 5)
print(2 ** 8)         # 256
print(10 ** 3)        # 1000

# Funziona anche con potenze decimali (radici!)
print(9 ** 0.5)       # 3.0 (radice quadrata di 9)
print(8 ** (1/3))     # 2.0 (radice cubica di 8)
```

## 3. Operatori e le loro priorità

Quando un'espressione contiene più operatori, Python deve decidere quale operazione eseguire per prima. Questo è determinato dalla **gerarchia delle priorità**.

### 3.1 La gerarchia delle priorità

Python segue queste regole (dalla priorità più alta alla più bassa):

1. **Parentesi**: `()`
2. **Esponente**: `**`
3. **Unario**: `+`, `-` (es: `-1`, `+3`)
4. **Moltiplicazione, Divisione, Modulo**: `*`, `/`, `//`, `%`
5. **Addizione, Sottrazione**: `+`, `-`

### Esempi pratici

```python
# Senza parentesi
print(2 + 3 * 4)              # 14 (non 20!)
# Spiegazione: 3 * 4 = 12, poi 2 + 12 = 14

# Con parentesi
print((2 + 3) * 4)            # 20
# Spiegazione: 2 + 3 = 5, poi 5 * 4 = 20

# Più operatori
print(10 - 2 * 3)             # 4 (non 24!)
# Spiegazione: 2 * 3 = 6, poi 10 - 6 = 4

print((10 - 2) * 3)           # 24
# Spiegazione: 10 - 2 = 8, poi 8 * 3 = 24

print(2 * 3 ** 2)             # 18 (non 36!)
# Spiegazione: 3 ** 2 = 9, poi 2 * 9 = 18

print((2 * 3) ** 2)           # 36
# Spiegazione: 2 * 3 = 6, poi 6 ** 2 = 36
```

### 3.2 Caso speciale: operatore unario dopo esponente

Un operatore unario a destra dell'esponente si lega più strettamente:

```python
print(4 ** -1)                # 0.25
# Spiegazione: -1 si lega prima a 4, quindi 4 ** (-1) = 1/4 = 0.25

print(-4 ** 2)                # -16 (non 16!)
# Spiegazione: 4 ** 2 = 16, poi -16
# Se vuoi (-4) ** 2 = 16, devi usare le parentesi!
print((-4) ** 2)              # 16
```

### 3.3 Espressioni complesse con parentesi

Le **sottoespressioni** tra parentesi vengono **sempre** calcolate per prime:

```python
print(15 - 1 * (5 * (1 + 2)))     # 0
# Passo per passo:
# 1 + 2 = 3
# 5 * 3 = 15
# 1 * 15 = 15
# 15 - 15 = 0

print(20 - (3 + 2) * 2)           # 10
# Passo per passo:
# 3 + 2 = 5
# 5 * 2 = 10
# 20 - 10 = 10
```

### 3.4 Associatività a destra: l'esponente

L'operatore di esponenziazione usa **binding a destra** (right-sided binding):

```python
print(2 ** 2 ** 3)                # 256 (non 64!)
# Spiegazione: viene calcolato da destra verso sinistra
# 2 ** 3 = 8
# 2 ** 8 = 256

# Se vuoi (2 ** 2) ** 3 = 64, usa le parentesi:
print((2 ** 2) ** 3)              # 64
```

Confronta con altri operatori (che usano binding a sinistra):

```python
print(10 - 5 - 2)                 # 3
# (10 - 5) - 2 = 5 - 2 = 3 (da sinistra a destra)

print(2 ** 3 ** 2)                # 512
# 2 ** (3 ** 2) = 2 ** 9 = 512 (da destra a sinistra!)
```

### 3.5 Tabella riassuntiva delle priorità

| Priorità | Operatori | Associatività | Esempio |
|----------|-----------|---------------|---------|
| 1 (massima) | `()` | - | `(2 + 3)` |
| 2 | `**` | Destra | `2 ** 3 ** 2` |
| 3 | `+`, `-` (unari) | - | `-5`, `+3` |
| 4 | `*`, `/`, `//`, `%` | Sinistra | `6 / 2 * 3` |
| 5 (minima) | `+`, `-` (binari) | Sinistra | `5 + 3 - 2` |
